```python
import streamlit as st
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import pickle
import seaborn as sns
import matplotlib.pyplot as plt

# Load the trained model and LabelEncoder
with open('model.pkl', 'rb') as model_file:
    model = pickle.load(model_file)

with open('label_encoder.pkl', 'rb') as encoder_file:
    l = pickle.load(encoder_file)

# Load the dataset for feature options
df = pd.read_csv('D:\\IQUBE\\INEURON\\mushrooms.csv')

# Feature columns (excluding the target column 'class')
feature_columns = [
    'cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor', 'gill-attachment', 
    'gill-spacing', 'gill-size', 'gill-color', 'stalk-shape', 'stalk-root', 
    'stalk-surface-above-ring', 'stalk-surface-below-ring', 'stalk-color-above-ring', 
    'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number', 'ring-type', 
    'spore-print-color', 'population', 'habitat'
]

# Streamlit Interface
st.title('Mushroom Classification')

st.write("Enter the feature values of the mushroom:")

# Create input fields for each feature
user_inputs = {}
for column in feature_columns:
    if df[column].dtype == 'object':  # For categorical features
        user_inputs[column] = st.selectbox(column, df[column].unique())
    else:  # For numerical features, if any
        user_inputs[column] = st.number_input(column, value=float(df[column].mean()))

# Convert the user inputs to the correct format
input_data = []
for column in feature_columns:
    if df[column].dtype == 'object':  # Handle categorical features
        try:
            # Use the previously fitted encoder (ensure it has all categories)
            input_data.append(l.transform([user_inputs[column]])[0])
        except ValueError:
            # Handle unseen labels by fitting a new LabelEncoder on all available categories
            temp_encoder = LabelEncoder()
            temp_encoder.fit(np.concatenate([df[column].values, [user_inputs[column]]]))
            input_data.append(temp_encoder.transform([user_inputs[column]])[0])
    else:  # Handle numerical features
        input_data.append(user_inputs[column])

# Convert the inputs into a DataFrame for prediction
input_df = pd.DataFrame([input_data], columns=feature_columns)

# Prediction button
if st.button('Classify'):
    prediction = model.predict(input_df)
    prediction_prob = model.predict_proba(input_df)[:, -1]  # Probability of the positive class
    
    # Show the result
    if prediction == 1:
        st.success(f"Prediction: Poisonous (Probability: {prediction_prob[0]:.2f})")
    else:
        st.success(f"Prediction: Edible (Probability: {prediction_prob[0]:.2f})")
```
